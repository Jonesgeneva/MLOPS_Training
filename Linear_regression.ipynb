```python
%matplotlib inline
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('smart_infrastructure_risk_dataset.csv')
df[['complaint_frequency', 'risk_score']].head()
```

```python
plt.xlabel('complaint_frequency')
plt.ylabel('risk_score')
plt.scatter(df['complaint_frequency'], df['risk_score'], color='red', marker='+')
plt.show()
```

```python
new_df = df[['complaint_frequency']]
new_df.head()
```

```python
risk_score = df.risk_score
risk_score.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, risk_score)
```

Predict risk score when complaint_frequency = 30

```python
reg.predict([[30]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

Y = m*X+b(m is coefficient and b is intercept)

```python
30 * 1.66287757 + 7.798826529672176
```

Save Model To a File Using Python Pickle

```python
import pickle
```

```python
with open('Linear regression','wb') as file:
    pickle.dump(reg,file)
```

Load Saved Model

```python
with open('Linear regression','rb') as file:
  mp = pickle.load(file)
```

```python
mp.intercept_
```

```python
mp.coef_
```

```python
mp.predict([[30]])
```

Save Trained Model Using joblib

```python
import joblib
```

```python
joblib.dump(reg, 'reg_joblib')
```

Load Saved Model

```python
mj = joblib.load('reg_joblib')
```
