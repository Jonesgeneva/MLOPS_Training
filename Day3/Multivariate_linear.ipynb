```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('smart_infrastructure_risk_dataset.csv')
df[['complaint_frequency','damage_size','previous_accidents','repair_delay_days','risk_score']].head(12)
```

Data Preprocessing: check median value of damage_size and fill any NA values

```python
df.damage_size.median()
```

```python
df.damage_size = df.damage_size.fillna(df.damage_size.median())
df[['complaint_frequency','damage_size','previous_accidents','repair_delay_days','risk_score']].head(12)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['complaint_frequency','damage_size','previous_accidents','repair_delay_days']], df.risk_score)
```

```python
reg.coef_
```

```python
reg.intercept_
```

Find the risk score of a complaint with complaint_frequency = 30,damage_size = 5.0, previous_accidents = 5, repair_delay_days = 90

```python
reg.predict([[30, 5.0, 5, 90]])
```
