```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('smart_infrastructure_risk_dataset.csv')
df['high_risk'] = (df['risk_level'] == 'high_risk').astype(int)
df[['damage_size','high_risk']].head()
```

```python
plt.scatter(df.damage_size, df.high_risk, marker='+', color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df[['damage_size']], df.high_risk, train_size=0.8, random_state=42
)
```

```python
X_test.head(6)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(damage_size):
    z = 2.691 * damage_size - 17.414
    y = sigmoid(z)
    return y
```

```python
damage_size = 2.0
prediction_function(damage_size)
```

```python
model.score(X_test, y_test)
```
