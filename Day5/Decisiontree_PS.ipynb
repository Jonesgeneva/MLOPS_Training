```python
import pandas as pd
```

```python
df = pd.read_csv('smart_infrastructure_risk_dataset.csv')
df.head()
```

```python
inputs = df.drop('risk_level', axis='columns')
target = df['risk_level']
```

```python
from sklearn.preprocessing import LabelEncoder

le_cat      = LabelEncoder()
le_severity = LabelEncoder()
le_rainfall = LabelEncoder()
le_traffic  = LabelEncoder()
```

```python
inputs['category_n']  = le_cat.fit_transform(inputs['category'])
inputs['severity_n']  = le_severity.fit_transform(inputs['severity'])
inputs['rainfall_n']  = le_rainfall.fit_transform(inputs['rainfall_level'])
inputs['traffic_n']   = le_traffic.fit_transform(inputs['traffic_density'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(
    ['complaint_id', 'complaint_text', 'location',
     'category', 'severity', 'rainfall_level', 'traffic_density'],
    axis='columns'
)
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```
